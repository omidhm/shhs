```python
import numpy as np
import matplotlib.pyplot as plt

# Example confusion matrix data
confusion_matrix = np.array([[45, 3, 2],
                             [6, 55, 1],
                             [0, 8, 48]])

# Define class labels
labels = ['Class A', 'Class B', 'Class C']

# Create figure and axis
fig, ax = plt.subplots()

# Create color map
cmap = plt.cm.Blues

# Plot the confusion matrix
im = ax.imshow(confusion_matrix, interpolation='nearest', cmap=cmap)

# Add color bar
cbar = ax.figure.colorbar(im, ax=ax)

# Set labels, title, and ticks
ax.set(xticks=np.arange(confusion_matrix.shape[1]),
       yticks=np.arange(confusion_matrix.shape[0]),
       xticklabels=labels, yticklabels=labels,
       title='Performance Confusion Matrix',
       ylabel='True Label',
       xlabel='Predicted Label')

# Rotate and align the tick labels
plt.setp(ax.get_xticklabels(), rotation=45, ha='right', rotation_mode='anchor')

# Loop over data dimensions and create text annotations
for i in range(confusion_matrix.shape[0]):
    for j in range(confusion_matrix.shape[1]):
        ax.text(j, i, format(confusion_matrix[i, j], 'd'),
                ha="center", va="center",
                color="white" if confusion_matrix[i, j] > confusion_matrix.max() / 2. else "black")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example confusion matrix data
confusion_matrix = np.array([[90, 10],
                             [15, 85]])

# Compute overall accuracy percentage
overall_accuracy = np.trace(confusion_matrix) / np.sum(confusion_matrix) * 100

# Compute percentages
confusion_matrix_percent = confusion_matrix.astype('float') / np.sum(confusion_matrix) * 100

# Define class labels
labels = ['Class 0', 'Class 1']

# Create figure and axis
fig, ax = plt.subplots()

# Create color map
cmap = plt.cm.Blues

# Plot the confusion matrix
im = ax.imshow(confusion_matrix_percent, interpolation='nearest', cmap=cmap)

# Add color bar
cbar = ax.figure.colorbar(im, ax=ax)

# Set labels, title, and ticks
ax.set(xticks=np.arange(confusion_matrix_percent.shape[1]),
       yticks=np.arange(confusion_matrix_percent.shape[0]),
       xticklabels=labels, yticklabels=labels,
       title='Performance Confusion Matrix\nOverall Accuracy: {:.2f}%'.format(overall_accuracy),
       ylabel='True Label',
       xlabel='Predicted Label')

# Rotate and align the tick labels
plt.setp(ax.get_xticklabels(), rotation=45, ha='right', rotation_mode='anchor')

# Loop over data dimensions and create text annotations with percentages
for i in range(confusion_matrix_percent.shape[0]):
    for j in range(confusion_matrix_percent.shape[1]):
        ax.text(j, i, f'{confusion_matrix_percent[i, j]:.2f}%',
                ha="center", va="center",
                color="white" if confusion_matrix_percent[i, j] > 0.5 else "black",
                fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
