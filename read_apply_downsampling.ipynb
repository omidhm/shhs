```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import decimate

# Load the EEG signal
Signal_EEG = np.load("D:\\Directory_D_Desk_jup\\Raw_npy\\Signal_EEG\\0\\shhs1-200033.npy")

# Define the original sampling rate and the duration to plot (30 seconds)
original_sampling_rate = 125
duration = 30  # seconds

# Define downsampling factors
downsampling_factors = [2, 4, 25]  # For example, downsample by factors of 2, 4, and 8

# Extract the portion of the signal to plot
num_samples_to_plot = duration * original_sampling_rate
signal_to_plot = Signal_EEG[num_samples_to_plot:2*num_samples_to_plot]

# Time vector for the original signal
time_vector_original = np.arange(num_samples_to_plot) / original_sampling_rate

# Plot original signal
plt.figure(figsize=(15, 10))
plt.subplot(4, 1, 1)
plt.plot(time_vector_original, signal_to_plot, label='Original Signal')
plt.title('Original Signal (125 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

# Plot downsampled signals
for i, factor in enumerate(downsampling_factors):
    downsampled_signal = decimate(signal_to_plot, factor)
    downsampled_time_vector = np.arange(len(downsampled_signal)) / (original_sampling_rate / factor)
    
    plt.subplot(4, 1, i + 2)
    plt.plot(downsampled_time_vector, downsampled_signal, label=f'Downsampled Signal (Factor {factor})')
    plt.title(f'Downsampled Signal (Factor {factor})')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()

plt.tight_layout()
plt.show()
```
